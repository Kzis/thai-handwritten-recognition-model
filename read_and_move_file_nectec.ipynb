```python
import glob
import os.path
from os import path
from shutil import copyfile
import pandas as pd
```

```python
def check_exist_path(file_path):
    print(path.exists(file_path))
    return path.exists(file_path)
```

```python
def create_folder(file_path):
    print("Create path {}".format(file_path))
    os.makedirs(file_path)
```

```python
def copy_file(source,destination):
    print("Copy {} ".format(source))
    print("To {} ".format(destination))
    copyfile(source, destination)
```

```python
def get_current_path():
    return os.path.abspath(os.getcwd()) + "\\"
```

```python
def df_export_excel(list_filename,list_path,list_class):
    df = pd.concat([pd.DataFrame(list_path), pd.DataFrame(list_filename) , pd.DataFrame(list_class)], axis=1)
    df.columns = ['Path', 'File','Class']
    df.to_csv(get_current_path() + "\\" +  "df_output.csv",index= False , encoding="utf-8-sig")
```

```python
list_filename = []
list_path = []
list_class = []

current_path = get_current_path()
read_path = "./*.jpg"

# Read image all directory
for file in glob.glob(read_path):
    file_str = str(file)
    file_opt = file_str.split("_")
    folder_name = file_opt[0].replace(".\\","")
    file_name = file_opt[1]
    
    if not check_exist_path(current_path + folder_name):
        create_folder(folder_name)
    
    source =  current_path + file_str
    destination =  current_path + "\\" +  folder_name + "\\" + (folder_name + "_" + file_name)
    copy_file(source,destination)
    
    list_filename.append((folder_name + "_" + file_name))
    list_path.append(destination)
    list_class.append(folder_name)

df_export_excel(list_filename,list_path,list_class)
```
